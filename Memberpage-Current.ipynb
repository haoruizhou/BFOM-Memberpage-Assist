```python
import pandas as pd
memberList = pd.read_csv('memberListCurrent.csv', sep=';')
memberList = memberList.fillna('')
print(memberList)
```

```python
# define
name = ""
# this is the text shown if image doesn't load
# example: Hally Zhou
# case NOT sensitive
fileName = ""
# this is the file name of the member's photo
# example: HallyZhou.jpg
# case SENSITIVE, practice
accessibleText = ""
# this text is for people who have impaired vision, when they use screen reader, this is the text they will hear
# example: Vice President Hally Zhou
# case NOT sensitive
parts = ""
# example: Bass
# case NOT sensitive
quote = ""
# example: "some quote"
# case NOT sensitive
position = ""
# example = "Vice President"
# case NOT sensitive
nameAndPronouns = ""
# this is your name and pronouns
# example: Hally Zhou (he/they)
# case NOT sensitive
xYearMajorMinor = ""
# example: 3rd Year Geography and Psychological & Brain Sciences
# case SENSITIVE
oneMember = ""
fourMember = ""
nameHeader = ""
# a string of 4 or less names that shows who's included in this code block

output = ""
```

```python
def oneMember(name, fileName, accessibleText, parts, quote, nameAndPronouns, position, xYearMajorMinor):

    temp = """
                <!-- NEW MEMBER HERE -->

                <div class="col-md-3 col-sm-3" style="margin-bottom:15px">
                    <div class="team-member">
                        <div class="team-img">
                            <img src="img/members/{fileName}" id = "{name}" alt="{accessibleText}" class="img-responsive">
                        </div>
                        <div class="team-hover">
                            <div class="desk">
                                <h4>{parts}</h4>
                                <p>{quote}</p>
                            </div>
                        </div>
                    </div>
                    <div class="team-title">
                        <h4>{nameAndPronouns}</h4>
                        <h5>{position}</h5>
                        <span>{xYearMajorMinor}</span>
                    </div>
                </div>

""".format(fileName = fileName, name = name, accessibleText = accessibleText, parts = parts, quote = quote, nameAndPronouns= nameAndPronouns, position = position, xYearMajorMinor = xYearMajorMinor)
    return temp

def templateOutsideBracket(nameHeader, fourMember):
    temp = """
                <!-- USE THIS TEMPLATE FOR EACH NEW ROW. REPLACE NAME/MAJOR/PARTS/POSITION/QUOTE PLACEHOLDER -->

                <!-- NEW ROW EVERY 4 MEMBERS -->
                <!-- LIST NAMES HERE FOR QUICK REFERENCE -->
                <!-- {nameHeader} -->
                <div class = "row" style = "margin-top:50px;">
                    {memberInfo}
                </div>
                <!-- END OF A FOUR PEOPLE BLOCK -->

    """.format(nameHeader = nameHeader, memberInfo = fourMember)

    return temp
```

```python
counter = 0

for i in range (len(memberList.index)):
    counter += 1
    print(memberList['name'].values[i])
    name = memberList['name'].values[i]
    nameHeader = str(nameHeader) + " " + str(name)
    nameAndPronouns = memberList['nameAndPronouns'].values[i]
    fileName = memberList['fileName'].values[i]
    accessibleText = memberList['accessibleText'].values[i]
    parts = memberList['parts'].values[i]
    quote = memberList['quote'].values[i]
    position = memberList['position'].values[i]
    xYearMajorMinor = memberList['xYearMajorMinor'].values[i]

    temp = oneMember(name, fileName, accessibleText, parts, quote, nameAndPronouns, position, xYearMajorMinor)
    fourMember = fourMember + temp

    if counter == 4 or i == len(memberList.index)-1:
        # 4 members is one row
        temp = templateOutsideBracket(nameHeader, fourMember)
        output = output + temp
        counter = 0
        fourMember = ""
        nameHeader = ""

    print(i)
```

```python
print(output)

with open('outputCurrent.txt', 'w') as f:
    f.write(output)
```
